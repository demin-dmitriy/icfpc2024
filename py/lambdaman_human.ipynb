```python
def save_compact_solution(task_id: int, compact_solution: ICFP) -> Path:
    assert isinstance(compact_solution, ICFP), type(ICFP)

    content = concat(
        ICFP.str(f'solve lambdaman{task_id} '),
        compact_solution,
    ).value

    destination = Path(f'solutions/lambdaman/compact/{task_id}_{len(content)}')
    if destination.exists():
        print(f'{destination} exists')

        #assert content == destination.read_text()

    destination.write_text(content)
    return destination


def post_solution(task_id: int, solution: str):
    save_to = Path(f'solutions/lambdaman/reported/{task_id}_{len(solution)}')

    if save_to.exists():
        print(f'already reported? {save_to}')
        return

    response = com.com(solution, True)
    save_to.write_text(response[0])

    print(response)
```

```python
def repeat_expr(expr: ICFP, times: int):
    return apply_lambda(
        apply_lambda(
            vf,
            expr
        ),
        ICFP.int(times)
    )

def repeat_str(s: str, times: int):
    return repeat_expr(ICFP.str(s), times)
```

```python
def merge_repeated_exprs(exprs: list[ICFP]):
    repeat_str_x_times = ly + y_combinator(f,
        lx + op_if(op_equal(vx, ICFP.int(0)),
            ICFP.str(''),
            concat(
                apply_lambda(vf, op_sub(vx, ICFP.int(1))),
                vy,
            )
        )
    )

    return let(f, repeat_str_x_times,
        concat(*exprs)
    )
```

```python
def find_patterns(s: str) -> list[tuple[str, int]]:
    parts = [] # (d, count)
    for d in s:
        if len(parts) == 0 or parts[-1][0] != d:
            parts.append([d, 1])
        else:
            parts[-1][1] += 1

    parts2 = []
    i = 0
    while i < len(parts):
        el = parts[i]

        if i + 3 < len(parts) and parts[i] == parts[i + 2] and parts[i + 1] == parts[i + 3]:
            pattern = parts[i][0] * parts[i][1] + parts[i + 1][0] * parts[i + 1][1]
            repeated = 1
            i += 2
            while i + 1 < len(parts) and pattern == parts[i][0] * parts[i][1] + parts[i + 1][0] * parts[i + 1][1]:
                repeated += 1
                i += 2

            parts2.append([pattern, repeated])

        parts2.append(parts[i])
        i += 1

    parts3 = []
    i = 0
    while i < len(parts2):
        if i + 1 == len(parts2):
            parts3.append(parts2[i])
            i+=1
        elif parts2[i][1] == 1 and parts2[i + 1][1] == 1:
            parts3.append([parts2[i][0] + parts2[i + 1][0], 1])
            i+=2
        elif parts2[i][1] == 1 and parts3[-1][1] == 1:
            parts3[-1][0] += parts2[i][0]
            i+=1
        else:
            parts3.append(parts2[i])
            i+=1

    return parts3
```

```python
ship_patterns = find_patterns(ship24)
ship_patterns[:5]
```

```python
def reproduce_pattern(pattern: list[tuple[str, int]]) -> ICFP:
    repeat_str_x_times = ly + y_combinator(f,
        lx + op_if(op_equal(vx, ICFP.int(0)),
            ICFP.str(''),
            concat(
                apply_lambda(vf, op_sub(vx, ICFP.int(1))),
                vy,
            )
        )
    )

    def select_encode(string, repeate_times):
        with_lambda = apply_lambda(
            apply_lambda(
                vf,
                ICFP.str(string)
            ),
            ICFP.int(repeate_times)
        )

        without_lambda = ICFP.str(string * repeate_times)

        if len(with_lambda.value) < len(without_lambda.value):
            return with_lambda

        return without_lambda


    return let(f, repeat_str_x_times,
        concat(*[
            select_encode(string, repeate_times)
            for string, repeate_times in pattern
        ])
    )
```
